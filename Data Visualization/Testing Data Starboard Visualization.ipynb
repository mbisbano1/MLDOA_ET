```python
#import the needed libraries pandas for dataframes and pyplot to plot
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#import the csv into a pandas dataframe type
data=pd.read_csv('Test_Stbd1.csv')
```

```python
data
```

```python
# extract the data
doa=data.DOA
sample_num=data.iloc[:,1] # Gets Sample# column
c=data.C #speed of sound
ping_num=data.iloc[:,0] # Gets Ping Numbers
i1=data.I1
i2=data.I2
i3=data.I3
i4=data.I4
i5=data.I5
i6=data.I6
i7=data.I7
i8=data.I8
i9=data.I9
i10=data.I10
q1=data.Q1
q2=data.Q2
q3=data.Q3
q4=data.Q4
q5=data.Q5
q6=data.Q6
q7=data.Q7
q8=data.Q8
q9=data.Q9
q10=data.Q10
```

```python
#plot i1 and q1 versus DOA (1st 10000 pts)
#this shows the releationship between I and Q data and DOA
plt.plot(i1[1:10000],doa[1:10000],'o')
plt.plot(q1[1:10000],doa[1:10000], 'o')
plt.title('I1 and Q1 VS DOA')
plt.xlabel('I1 and Q1 data')
plt.ylabel('DOA (degrees)')
plt.legend(['I1','Q1'])
plt.show()
```

```python
#plot i1 and q1 versus DOA (2nd 10000 pts)
#this shows the releationship between I and Q data and DOA
plt.plot(i1[10000:20000],doa[10000:20000],'o')
plt.plot(q1[10000:20000],doa[10000:20000], 'o')
plt.title('I1 and Q1 VS DOA')
plt.xlabel('I1 and Q1 data')
plt.ylabel('DOA (degrees)')
plt.legend(['I1','Q1'])
plt.show()
```

```python
#plot i2 and q2 versus DOA (1st 10000 pts)
#this shows the releationship between I and Q data and DOA
plt.plot(i2[1:10000],doa[1:10000],'o')
plt.plot(q2[1:10000],doa[1:10000], 'o')
plt.title('I2 and Q2 VS DOA (1st 10,000 pts)- STARBOARD')
plt.xlabel('I2 and Q2 data')
plt.ylabel('DOA (degrees)')
plt.legend(['I2','Q2'])
plt.show()
```

```python
#plot i2 and q2 versus DOA (2nd 10000 pts)
#this shows the releationship between I and Q data and DOA
plt.plot(i2[10000:20000],doa[10000:20000],'o')
plt.plot(q2[10000:20000],doa[10000:20000], 'o')
plt.title('I2 and Q2 VS DOA (2nd 10000 pts)- STARBOARD')
plt.xlabel('I2 and Q2 data')
plt.ylabel('DOA (degrees)')
plt.legend(['I2','Q2'])
plt.show()

print("this shows that there is a strong relationship between both the i and q data with the direction of arrival")
print("no matter where I sample the I and Q data, the same patterns occur.")
print("if this is truly the case, then MLDOA will have an easier time learning I believe")
```

```python
#show the realationship between DOA and concurrent samples (1st 10,000 pts)
plt.plot(sample_num[0:4284],doa[0:4284],'o')
plt.title('sample number VS DOA')
plt.xlabel('sample number')
plt.ylabel('DOA (degrees)')
plt.show()
```

```python
fig=plt.figure()
ax1=fig.add_subplot(211)
ax2=fig.add_subplot(212)

#fig.suptitle('Plots with differnet Samples')

ax1.plot(i2[1:10000],doa[1:10000],'o')
ax1.plot(q2[1:10000],doa[1:10000], 'o')
ax1.set_xlabel('I2 and Q2 Data')
ax1.set_ylabel('DOA (degrees)')
ax1.title.set_text('I2 and Q2 VS DOA (1st 10,000 pts)- STARBOARD')
ax1.legend(['I2','Q2'])

ax2.plot(i2[10000:20000],doa[10000:20000],'o')
ax2.plot(q2[10000:20000],doa[10000:20000], 'o')
ax2.set_ylabel('DOA (degrees)')
ax2.set_xlabel('I2 and Q2 Data')
ax2.title.set_text('I2 and Q2 VS DOA (2nd 10,000 pts)- STARBOARD')
ax2.legend(['I2','Q2'])

fig.tight_layout()
plt.show()

print("this shows that there is a strong relationship between both the i and q data with the direction of arrival")
print("no matter where I sample the I and Q data, the same patterns occur.")
print("if this is truly the case, then MLDOA will have an easier time learning I believe")
```
